```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
trafic = pd.read_table('report1.txt')
```

```python
trafic
```

```python
#중구 데이터 , 발생건수만 데이터 뽑기
junggu = trafic[trafic.자치구별 == "중구"][trafic.구분 == "발생건수"].iloc[:,-12:] # 뒤에서 12개 빼야하니까 -12
junggu
junggu = junggu.astype('int') # 문자열에서 정수로 변환
```

```python
# #한글 처리 코드
# !apt -y install fonts-nanum > /dev/null
# !fc-cache -fv > /dev/null
# !rm -rf ~/.cache/matplotlib > /dev/null
plt.rc('font', family='NanumBarunGothic')
```

```python
plt.plot(junggu.values[0]) # 2차원에서 1차원으로 바꿔줌
plt.title("2019년 중구 월별 교통사고 발생건수")
plt.xlabel("월")
plt.ylabel("발생건수")
plt.grid()
plt.ylim(0,120)
plt.xticks(range(12), ['1월', '2월', '3월', '4월', '5월', '6월', '7월', '8월', '9월', '10월', '11월', '12월'])
plt.show() #프레임 상태로 데이터 시각화 안됨 / values로 데이터 추출하고 1차원으로 만들어서 그래프
```
