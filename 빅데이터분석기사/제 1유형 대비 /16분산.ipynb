# 주어진 데이터 셋에서 f2가 0값인 데이터를 age를 기준으로 오름차순 정렬하고
# 앞에서 부터 20개의 데이터를 추출한 후 
# f1 결측치(최소값)를 채우기 전과 후의 분산 차이를 계산하시오 (소수점 둘째 자리까지)

```python
import pandas as pd
import numpy as np
df = pd.read_csv('/Users/jochaeyeon/Downloads/basic1.csv')
df.head()
```

```python
df = df[df["f2"] == 0]
df = df.sort_values(by="age", ascending=False)
top_20 = df.head(20)
top_20
```

```python
df_var = top_20["f1"].var()
df_var
```

```python
min = top_20["f1"].min()
top_20["f1"] = top_20["f1"].fillna(min)
df_var2 = df["f1"].var()
df_var2
```

```python
print(round(df_var2-df_var,2))
```
