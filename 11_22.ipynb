```python
# 피마 인디언 데이터 생성
pima = pd.read_csv('pima-indians-diabetes (1).csv', header=None)
pima.head()
```

```python
#칼럼 이름 변경
pima.columns = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
```

```python
# 데이터 정보 확인
pima.info()
```

```python
pima.describe()
```

```python
pima.isnull().sum()
```

```python
#독립변수와 종속변수 분리
y = pima['class']
X = pima.drop('class', axis=1)
```

```python
# x 분포 확인
plt.boxplot(X)
plt.show()

# 현재 5 변수만 이상치 많음 ... 
```

```python
# scaling ( 0과 1 사이)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X)
scaled_X = scaler.fit_transform(X)
scaled_X
```

```python
# 스케일링 된 후 분포확인
plt.boxplot(scaled_X)
plt.show()
```

```python
#Splitting the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(scaled_X, y, test_size=0.2, random_state=0)
```

```python
y_train = pd.DataFrame(y_train)
y_train
```

```python
sns.countplot(data= y_train , x= 'class', hue='class')
plt.show() # 1번이 당뇨병 있는 사람, 0번이 업는 사람
```

```python
#resampling
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=0)
X_resample, y_resample = smote.fit_resample(X_train, y_train)
```

```python
#resampling 후 데이터 분포 확인
sns.countplot(data= y_resample , x= 'class', hue='class')
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
models = [DecisionTreeClassifier(),RandomForestClassifier(),LinearDiscriminantAnalysis(),KNeighborsClassifier(),SVC()]
cv_list = []
for model in models:
     model.fit(X_resample, y_resample) #학습
     cv_list.append(cross_val_score(model, X_test, y_test, cv=10, scoring='accuracy'))
cv_list
```

```python
plt.boxplot(cv_list)
plt.xticks([1,2,3,4,5],['DT','RF','LDA','KNN','SVM'])
plt.show()
```

```python
tab = pd.DataFrame(cv_list, index=['DT', 'RF', 'LDA', 'KNN', 'SVM'])
tab = tab.T 
```

```python
#교차검증 요약
summary = pd.concat([tab.mean(), tab.std()], axis=1)
summary.columns = ['Mean', 'STD']
summary
```
