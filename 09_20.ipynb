```python
#전체주문 아이템의 수
number = print(chipo.quantity.sum())
```

```python
#데이터 셋 안에 있는 수익 ?
print((chipo['quantity']*chipo['item_price']).sum())
```

```python
# 가장 많이 주문된 아이템
print('가장 많이 주문된 아이템은?')
print(chipo.groupby('item_name')['quantity'].sum().idxmax())
print('\n')

# 가장 많이 주문된 아이템은 몇 개나 주문됐나요? (quantity의 합계)
print('가장 많이 주문된 아이템은 몇개나 주문됐나요?')
print(chipo.groupby('item_name').size().max())
print('\n')

# choice_description 컬럼에서 가장 많이 주문된 아이템 (quantity의 합계 기준)
print('choice_description 컬럼에서 가장 많이 주문된 아이템은?')
print(chipo.groupby('choice_description')['quantity'].sum().idxmax())
print('\n')

# choice_description에 따라 Groupby 된 것 중 가장 카운트가 많은 것
print('choice_description에 따라 Groupby 된 것 중 가장 카운트가 많은 것은?')
print(chipo['choice_description'].value_counts().max())
print('\n')

# 얼마나 많은 다른 종류의 아이템이 팔렸나요? (고유 item_name의 수)
print('얼마나 많은 다른 종류의 아이템이 팔렸나요?')
print(chipo['item_name'].nunique())
```
